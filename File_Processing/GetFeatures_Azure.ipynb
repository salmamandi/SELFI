```python
import os
import io
import json
import glob
from azure.cognitiveservices.vision.face import FaceClient
from msrest.authentication import CognitiveServicesCredentials
import requests
from PIL import Image, ImageDraw, ImageFont
import csv
import pandas as pd
```

```python
# provide credential 
credential=json.load(open('AzureConfig.json'))
API_KEY=credential['API_KEY']
ENDPOINT=credential['EndPoint']
face_client=FaceClient(ENDPOINT,CognitiveServicesCredentials(API_KEY))
```

```python
csv_columns=['FileName','pupil_left', 'pupil_right', 'nose_tip', 'mouth_left', 'mouth_right', 
             'eyebrow_left_outer', 'eyebrow_left_inner', 'eye_left_outer', 'eye_left_top', 'eye_left_bottom', 
             'eye_left_inner', 'eyebrow_right_inner', 'eyebrow_right_outer', 'eye_right_inner', 'eye_right_top',
             'eye_right_bottom', 'eye_right_outer', 'nose_root_left', 'nose_root_right', 'nose_left_alar_top', 
             'nose_right_alar_top', 'nose_left_alar_out_tip', 'nose_right_alar_out_tip', 'upper_lip_top', 
             'upper_lip_bottom', 'under_lip_top', 'under_lip_bottom']
feature_dict=[]
user_no=12
# path for image file
search_path="Data/P5/saved_images"

for root, directory, files in os.walk(search_path):
       for image_file_name in files: 
            image_file_path=search_path+"/"+image_file_name
            with open(image_file_path,'rb') as image_file:
                response_detected_faces=face_client.face.detect_with_stream(image_file,return_face_landmarks=True,
                                                                            detection_model='detection_03',
                                                                            recognition_model='recognition_04',
                                                                            return_recognition_model=True)
                #print(response_detected_faces)
                if not response_detected_faces:
                    print("definitly not")
                else:
                    for i in range(len(response_detected_faces)):
                        person1=response_detected_faces[i]
                        row_dict=dict()
                        row_dict["FileName"]="user_"+str(user_no)+"/"+image_file_name
                        row_dict['pupil_left']=vars(person1.face_landmarks.pupil_left)
                        row_dict['pupil_right']=vars(person1.face_landmarks.pupil_right)
                        row_dict['nose_tip']=vars(person1.face_landmarks.nose_tip)
                        row_dict['mouth_left']=vars(person1.face_landmarks.mouth_left)
                        row_dict['mouth_right']=vars(person1.face_landmarks.mouth_right)
                        row_dict['eyebrow_left_outer']=vars(person1.face_landmarks.eyebrow_left_outer)
                        row_dict['eyebrow_left_inner']=vars(person1.face_landmarks.eyebrow_left_inner)
                        row_dict['eye_left_outer']=vars(person1.face_landmarks.eye_left_outer)
                        row_dict['eye_left_top']=vars(person1.face_landmarks.eye_left_top)
                        row_dict['eye_left_bottom']=vars(person1.face_landmarks.eye_left_bottom)
                        row_dict['eye_left_inner']=vars(person1.face_landmarks.eye_left_inner)
                        row_dict['eyebrow_right_inner']=vars(person1.face_landmarks.eyebrow_right_inner)
                        row_dict['eyebrow_right_outer']=vars(person1.face_landmarks.eyebrow_right_outer)
                        row_dict['eye_right_inner']=vars(person1.face_landmarks.eye_right_inner)
                        row_dict['eye_right_top']=vars(person1.face_landmarks.eye_right_top)
                        row_dict['eye_right_bottom']=vars(person1.face_landmarks.eye_right_bottom)
                        row_dict['eye_right_outer']=vars(person1.face_landmarks.eye_right_outer)
                        row_dict['nose_root_left']=vars(person1.face_landmarks.nose_root_left)
                        row_dict['nose_root_right']=vars(person1.face_landmarks.nose_root_right)
                        row_dict['nose_left_alar_top']=vars(person1.face_landmarks.nose_left_alar_top)
                        row_dict['nose_right_alar_top']=vars(person1.face_landmarks.nose_right_alar_top)
                        row_dict['nose_left_alar_out_tip']=vars(person1.face_landmarks.nose_left_alar_out_tip)
                        row_dict['nose_right_alar_out_tip']=vars(person1.face_landmarks.nose_right_alar_out_tip)
                        row_dict['upper_lip_top']=vars(person1.face_landmarks.upper_lip_top)
                        row_dict['upper_lip_bottom']=vars(person1.face_landmarks.upper_lip_bottom)
                        row_dict['under_lip_top']=vars(person1.face_landmarks.under_lip_top)
                        row_dict['under_lip_bottom']=vars(person1.face_landmarks.under_lip_bottom)
                        #print(row_dict)    
                        for item in row_dict.keys():
                            if 'additional_properties' in row_dict[item]:
                                del row_dict[item]['additional_properties']
                        #final row dict        
                        #print(row_dict) 
                        feature_dict.append(row_dict)

# write features in the file 
csv_file="Data/P5/user_"+str(user_no)+".csv"
try:
    with open(csv_file, 'w') as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
        writer.writeheader()
        for data in feature_dict:
            #print(data)
            writer.writerow(data)
except IOError:
    print("I/O error")    
```
